```python
from collections import defaultdict

csv_lines = [
    "John.IT.50000",
    "Mary, IT,55000",
    "Alice.HR,48000",
    "Bob,HR,52000"
]

def parse_line(line):
    if ',' in line:
        parts = [x.strip() for x in line.split(',')]
    else:
        parts = [x.strip() for x in line.split('.')]
    if len(parts) == 3:
        return parts[0], parts[1], parts[2]
    elif len(parts) == 2:
        name_dept = parts[0].split('.')
        return name_dept[0], name_dept[1], parts[1]
    else:
        raise ValueError("Line format not recognized: " + line)
salary_by_dept = defaultdict(list)

for line in csv_lines:
    name, dept, salary = parse_line(line)
    salary_by_dept[dept].append(float(salary))

avg_salary_by_dept = {ept: sum(salaries)/len(salaries) for dept, salaries in salary_by_dept.items()}

print("Average Salary by Department:")
for dept, avg_salary in avg_salary_by_dept.items():
    print(f"{dept}: {avg_salary}")
```
