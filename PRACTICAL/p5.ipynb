```python
employees = [
    {"name": "Alice", "salary": 55000, "department": "HR"},
    {"name": "Bob", "salary": 45000, "department": "Finance"},
    {"name": "Charlie", "salary": 70000, "department": "IT"},
    {"name": "David", "salary": 60000, "department": "Marketing"}
]

# Sorting the list by salary (ascending)
sorted_employees = sorted(employees, key=lambda emp: emp["salary"])

# Display the sorted list
print("Employees sorted by salary:")
for emp in sorted_employees:
    print(emp)
```
