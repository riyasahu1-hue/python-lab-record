```python
# Loop up to 5 and print numbers from 1 to i in each row
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=' ')
    print()
```

```python
# Number of rows
rows = 5

for i in range(1, rows + 1):
    print('*' * (2*i - 1))
```

```python
# Diamond pattern

rows = 5  # height of the top half

# Top half
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2*i - 1))

# Bottom half
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i) + "*" * (2*i - 1))
```

```python
n=5
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("*", end="")
        else:
            print("", end="")
            print()
```
