```python

user_list = input("Enter items separated by spaces: ").split()

# Take the item to search for
item = input("Enter the item to search for: ")

# Search for the item
if item in user_list:
    position = user_list.index(item)  # Get index
    print(f"Item found at position {position}")  # 0-based index
else:
    print("Item not found.")
```
