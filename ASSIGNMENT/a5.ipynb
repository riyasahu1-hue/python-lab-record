```python
def reverse_vowels(s):
    vowels = "aeiouAEIOU"
    s = list(s)  # Convert string to list for easy modification

    # Collect all vowels in the string
    vowel_list = [char for char in s if char in vowels]

    # Reverse the vowel list
    vowel_list.reverse()

    # Replace vowels in original string with reversed vowels
    index = 0
    for i in range(len(s)):
        if s[i] in vowels:
            s[i] = vowel_list[index]
            index += 1

    return "".join(s)


# Taking input
string = input("Enter a string: ")
result = reverse_vowels(string)

print("String after reversing vowels:", result)
```
